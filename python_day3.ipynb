```python
'''🚗 Problem: “Car Rental Cost Estimator” 🎯 Problem Statement

Create a function that calculates rental cost.

📌 Requirements:

Parameters:

Number of days

Cost per day

Default insurance charge per day = 100

Calculate total cost including insurance.

If rental days > 7, give 5% discount on total.

💡 Sample Call: calculate_rent(5, 1000) 📤 Expected Output: Days: 5 Rent Cost: 5000 Insurance: 500 Total Payable: 5500
💡 Sample 2: calculate_rent(10, 1000) Days: 10 Rent Cost: 10000 Insurance: 1000 Discount Applied: 550 Final Payable: 10450'''
```

```python
def car_rental_cost_estimator(number_of_days,rental, default_insurance_per_day=100):
  discount=0
  if number_of_days>7:
    discount=0.05
  total= number_of_days*rental + number_of_days*default_insurance_per_day
  return total-(discount*total)

 #number_of_days = map(int,input("enter number of days:").split())
  #cost_per_day = int(input("enter "))
```

```python
car_rental_cost_estimator(5,1000,100)
```

```python
'''Problem: “Startup Expense Tracker” 🎯 Problem Statement

Create a function that calculates total business expenses.

📌 Requirements:

Accept business name (mandatory parameter)

Accept any number of expenses using *kwargs

Print each expense category with amount

Print total expense

If no expenses are given, print "No expenses recorded." office=15000 staff=150000 food=10000 tea=2000 ads=100000'''
```

```python
def startup_expens_tracker(name,*expenses):
    total_expense=0
    if len(expenses)==0:
        return f"{name}:No expenses recorded."
    else:
        print(f"Total expenses are:{len(expenses)}")
    for e in expenses:
        if e=="office":
            total_expense+=15000
        elif e=="staff":
            total_expense+=150000
        elif e=="food":
            total_expense+=10000
        elif e=="tea":
            total_expense+=2000
        elif e=="ads":
            total_expense+=100000
    return f"{name}:{total_expense}"
```
