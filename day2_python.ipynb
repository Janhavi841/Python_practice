```python
'''1️⃣ Ride-Sharing Daily Fare System
Business Context: Calculate total fare for multiple rides in a day.

Base Fare = ₹50
Per KM = ₹12
If distance > 20 km → 10% discount
Stop when distance = 0
Sample Input:
10
25
0
Sample Output:
Ride 1 Fare: ₹170
Ride 2 Fare: ₹324
Total Payable: ₹494'''

total_fare = 0
ride_number = 1

while True:
     distance = float(input("enter your ride distance:"))
     if distance == 0:
       break

     fare = 50 + (distance * 12)
     if distance > 20:
        fare = fare - (fare * 0.10)

print(f"Ride {ride_number} Fare: ₹{fare}")

total_fare += fare
ride_number += 1

print(f"Total Payable: ₹{total_fare}")

```

```python
'''2️⃣ Subscription Miss Tracker
Business Context: Cancel subscription if 2 consecutive missed payments.

Sample Input:

1
0
0
Sample Output:

Month 2: Payment Missed
Month 3: Payment Missed
Subscription Cancelled'''

missed = 0
month = 1

while True:
    payment = int(input("Enter payment (1 = Paid, 0 = Missed): "))

    if payment == 0:
        print(f"Month {month}: Payment Missed")
        missed += 1
    else:
        missed = 0   # reset if paid

    if missed == 2:
        print("Subscription Cancelled")
        break

    month += 10
```

```python
'''4️⃣ Startup Revenue Drop Alert
Logic: If revenue drops 3 consecutive months → Risk Alert.

Sample Input:

500000
450000
400000
380000
Sample Output:

Business Risk Alert in Month 4'''
previous = float(input("Enter Month 1 revenue: "))

drop_count = 0
month = 2

while True:
    current = float(input(f"Enter Month {month} revenue: "))

    if current < previous:
        drop_count += 1
    else:
        drop_count = 0

    if drop_count == 3:
        print(f"Business Risk Alert in Month {month}")
        break

    previous = current
    month += 1
```

```python
'''5️⃣ Loan EMI Eligibility System
Logic: EMI = ₹20000. If EMI > 40% of income → Not Eligible. Stop when -1 entered.

Sample Input:

60000
30000
-1
Sample Output:

Eligible
Not Eligible
Total Eligible: 1'''

emi = 20000
eligiblecount =0
while True:
  income = float(input("enter your income"))
  if income< 0:
        print("close")
        break
  elif emi > 0.4*(income):
         print("sorry not eligible")
  else:
        print("you are eligible for our LOAN servicee !!!")
        eligiblecount +=1

  print(f"eligibility count is : {eligiblecount}")
```

```python
'''6️⃣ Restaurant Table Billing System
Logic: Add items until 0. If bill > 5000 → 15% discount. Add 5% GST.

Sample Input:

2000
3000
7000
0
Sample Output:

Subtotal: ₹12000
Discount: ₹1800
GST: ₹510
Final Bill: ₹10710'''

sub_total=0
while True:
  user=float(input("enter your item prices :"))
  if user==0:
    print("--------------bill--------------")
    break
  sub_total +=user

discount=sub_total*0.15
gst=(sub_total-discount)*0.05
final_bill=sub_total+gst-discount

print(f"subtotal: {sub_total}")
print(f"discount: {discount}")
print(f"gst: {gst}")
print("------------------------------------")
print(f"final_bill: {final_bill}")
```

```python
'''7️⃣ Server Load Monitor
Logic: Track 24-hour requests. If requests > 10000 → Overload Alert. Count total overload hours.

Sample Output:

Hour 5: Overload
Hour 12: Overload
Total Overload Hours: 2'''

count = 0

for hrs in range (1,25):
  request = int(input(f"enter your reuests for hrs{hrs}"))

  if request > 10000:
     print(f"hrs{hrs}:overload")
     count+=1

print(f"total overload hrs:{count}")
```
